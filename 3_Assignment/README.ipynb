# Rock Critic Reviewer

**Rock Critic Reviewer** on komentorivipohjainen sovellus, joka hyödyntää OpenAI:n GPT-mallia generoimaan kolme erilaista arvostelua kullekin annettavalle albumille. Jokainen arvostelu edustaa yhtä seuraavista näkökulmista:

1. **Analyyttinen Puristi:**  
   Pureutuu levyn rakenteeseen, soittoon, sanoituksiin ja tuotantoon. Vertaa menneisiin suuruuksiin ja etsi armotta heikkouksia. Syväluotaavaa, kriittistä, historiaan ja teoriaan painottuvaa asiantuntijakirjoittamista.

2. **Trenditietoinen Kyynikko:**  
   Arvioi levyn ajankohtaisuutta, potentiaalista vaikutusta ja "cooliutta". Ole skeptinen hypen suhteen ja viittaa ajankohtaisiin ilmiöihin sarkastisesti.

3. **Vanhan Liiton Murisija:**  
   Arvioi levyn uskollisuutta "oikealle" rock-hengelle ja artistin aikaisemmalle tuotannolle. Ole epämuodollinen, puolueellinen, tunteellinen ja intohimoinen. Vihaa tai rakasta".

Arvostelut tulostetaan sekä komentoriville että tallennetaan Markdown-muotoon tiedostoon `album_reviews.md`, jotta ne ovat helposti luettavissa ja jaettavissa.

---

## Ominaisuudet

- **Interaktiivinen käyttöliittymä:**  
  Ohjelma kysyy albumin nimeä tai kuvausta ja jatkaa uusien arvostelujen generointia, kunnes syötteenä annetaan tyhjä rivi.

- **Monipuoliset arvostelut:**  
  Jokaisesta albumista generoidaan kolme erilaista arvostelua, jotka edustavat eri rockkriitikon näkökulmia.

- **Markdown-muotoinen ulostulo:**  
  Arvostelut tallennetaan tiedostoon, jossa käytetään Markdown-syntaksia, mikä parantaa luettavuutta ja mahdollistaa sisällön helpon jakamisen.

- **OpenAI:n GPT-malli:**  
  Käytössä on GPT-3.5-turbo, joka tarjoaa reippaan ja kriittisen luovan kirjoitustyylin kustannustehokkaasti.

## Asennus

### Vaaditut komponentit

- Python 3.6 tai uudempi
- OpenAI API-avain (aseta ympäristömuuttujaan `OPENAI_API_KEY`)
- Python-kirjasto `openai` (asennus: `pip install openai`)

### Vaiheet

1. **Projektin kloonaus:**  
   Lataa tai kloonaa tämä repository.

2. **Virtuaaliympäristön luominen (suositeltavaa):**
   ```bash
   python -m venv venv
   source venv/bin/activate    # Linux/macOS
   venv\Scripts\activate       # Windows

### Riippuvuuksien asennus:

pip install openai

### Ympäristömuuttujan asettaminen:

    Linux/macOS:

export OPENAI_API_KEY="oma_api_avain"

Windows:

set OPENAI_API_KEY=oma_api_avain

### Käyttö
Aja ohjelma komentoriviltä:

python rock_critic.py

Ohjelman kulku:

    Ohjelma kysyy:
    "Mikä albumi on analysoitavana?"

    Syötä albumin nimi tai kuvaus, jonka haluat arvostella.

    Ohjelma generoi kolme erilaista arvostelua (yksi kustakin näkökulmasta) ja tulostaa ne sekä näytölle että tallentaa ne tiedostoon album_reviews.md.

    Ohjelma kysyy:
    "Mitä albumia haluat seuraavaksi analysoida?"
    Syötä uusi albumi tai paina Enter lopettaaksesi ohjelman.

### Koodin Rakenne

    get_review:
    Funktio, joka vastaanottaa albumin nimen ja kriitikon näkökulman tiedot sekä kutsuu OpenAI:n API:a generoimaan napakan arvostelun.

    main:
    Pääsilmukka, joka hoitaa käyttäjän syötteet, arvostelujen generoinnin, tulostuksen sekä tiedostoon kirjoittamisen.

## Mukautus
